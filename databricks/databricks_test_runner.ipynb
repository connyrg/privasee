```python
import sys
import os
import subprocess

# Set paths
databricks_dir = "/Workspace/Users/Conny.GUNADI@suncorp.com.au/document_deidentification_tool/src/privasee/databricks"
parent_dir = "/Workspace/Users/Conny.GUNADI@suncorp.com.au/document_deidentification_tool/src/privasee"

# Setup environment variables
test_env = os.environ.copy()
test_env["AZURE_DOCUMENT_INTELLIGENCE_ENDPOINT"] = "https://test.cognitiveservices.azure.com/"
test_env["AZURE_DOCUMENT_INTELLIGENCE_KEY"] = "test-key-placeholder"
test_env["PYTHONDONTWRITEBYTECODE"] = "1"  # Prevent .pyc files
test_env["PYTHONPATH"] = parent_dir  # So 'databricks.model' imports work

print("Running tests...\n")
result = subprocess.run(
    [
        "python", "-m", "pytest", 
        "tests/", 
        "-v",
        "--tb=short",
        "--cov=databricks.model", 
        "--cov-report=term",
        "-p", "no:cacheprovider"  # Disable pytest cache
    ],
    capture_output=True,
    text=True,
    cwd=databricks_dir,
    env=test_env
)

print(result.stdout)
if result.stderr:
    print(result.stderr)
```

```python
import re

# Parse the output
if result.returncode == 0:
    print("✅ All tests passed!")
else:
    print(f"❌ Some tests failed (exit code: {result.returncode})")

# Extract coverage percentage
match = re.search(r'TOTAL.*?(\d+)%', result.stdout)
if match:
    coverage = int(match.group(1))
    print(f"\n📊 Code Coverage: {coverage}%")
    if coverage >= 90:
        print("🌟 Excellent coverage!")
    elif coverage >= 80:
        print("✅ Good coverage")
    else:
        print("⚠️  Consider adding more tests")
```
