```python
import requests
import json
import base64

endpoint_url = 'https://suncorp-dev.cloud.databricks.com/serving-endpoints/privasee_endpoint_local/invocations'
DATABRICKS_TOKEN = dbutils.secrets.get(scope='Conny.GUNADI@suncorp.com.au', key='DATABRICKS_TOKEN_DEV')

# Read and encode document
with open("/Volumes/datascience_dev_bronze_sandbox/ds_document_deidentification/test_documents/Sample GP document PDF.pdf", "rb") as f:
    doc_bytes = base64.b64encode(f.read()).decode()

field_definitions = [
    {
        "name": "claimant_name",
        "description": "The name of the claimant of a personal injury claims. The claimant is the person who is injured and lodging a compensation claim.",
        "strategy": "Black Out"
    },
    {
        "name": "incident_date",
        "description": "The date when an incident happened",
        "strategy": "Black Out"
    }
]

# Create request payload
payload = {
    "dataframe_records": [{
        "session_id": "session_123",
        "document_bytes_b64": doc_bytes,
        "document_filename": "original.pdf",
        "field_definitions_json": json.dumps(field_definitions)
    }]
}

# Invoke endpoint
response = requests.post(
    f"{endpoint_url}",
    headers={"Authorization": f"Bearer {DATABRICKS_TOKEN}"},
    json=payload
)

result = response.json()
```
